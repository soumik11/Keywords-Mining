```python
summary = (df
           .groupby(['displayLink'], as_index=False)
           .agg({'rank': ['count', 'mean']})
           .sort_values(('rank', 'count'), ascending=False)
           .assign(coverage=lambda df: df[('rank', 'count')]
                   .div(50)))


summary.columns = ['displayLink', 'count', 'avg_rank', 'coverage']
summary['displayLink'] = summary['displayLink'].str.replace('www.', '')
summary['avg_rank'] = summary['avg_rank'].round(1)
summary['coverage'] = summary['coverage'].mul(100).round(1).astype(str).add('%')
summary.head(50).reset_index(drop=True)
```

## co-relation between rankings and emojis

```python
import emoji
from collections import Counter

emojis_df=[]
for string in df['title']:
    my_str = str(string)
    for each in my_str:
        if each in emoji.UNICODE_EMOJI:
            emojis_df.append(each)
freq = dict(Counter(i for sub in emojis_df for i in set(sub))) 
sort_orders = sorted(freq.items(), key=lambda x: x[1], reverse=True)
sort_orders
```

### as we can see above for the title there is no emoji

```python
import emoji
from collections import Counter

emojis_df=[]
for string in df['snippet']:
    my_str = str(string)
    for each in my_str:
        if each in emoji.UNICODE_EMOJI:
            emojis_df.append(each)
freq = dict(Counter(i for sub in emojis_df for i in set(sub))) 
sort_orders = sorted(freq.items(), key=lambda x: x[1], reverse=True)
sort_orders
```
